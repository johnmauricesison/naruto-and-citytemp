```python
import pandas as pd

naruto = pd.read_csv('C:\\Users\\John Maurice Sison\\OneDrive\\Documents\\First Year Codes\\Datasets\\naruto.csv', encoding='cp1252')
```

Naruto Dataset

```python
naruto
```

1. Total Votes per Saga

```python
episode = naruto.groupby(by ='Saga').agg({'Votes':sum})
episode
```

2. Total Votes per Episode Type and Year


```python
episode = naruto.groupby(['Type','Year_launch'])[['Votes']].sum()
episode
```

3. Average Rate and Total Votes per Episode Type and Year

```python
votes = naruto.groupby(['Type' ,'Year_launch']).agg({'Rate':'mean','Votes':sum})
votes
```

4. Total Votes by Year. Sorted in Descending order based on Total Votes.

```python
numep = naruto.groupby(by = 'Year_launch').agg({'Votes':sum}).sort_values(by =['Votes'], ascending =False)
numep
```

5. Average Rate and Total Votes per Episode Type. Sorted in Descending order based on Average Rate.

```python
rates = naruto.groupby(by = 'Type').agg({'Rate':'mean','Votes':sum}).sort_values(by= ['Rate'], ascending = False)
rates
```

6. Average Rate and Total Votes per Saga. Sorted in Ascending order based on Average Rate.

```python
year = naruto.groupby(by = 'Saga').agg({'Rate':'mean','Votes':sum}).sort_values(by = 'Rate', ascending = True)
year
```

7. Top 20 Highest Average Rate per Episode Type, Year, and Saga. Sorted in Descending order based on Average Rate. 

```python
seven = naruto.groupby(by= ['Type','Year_launch','Saga']).agg({'Rate':'mean'}).sort_values(by = 'Rate', ascending = False).head(20)
seven
```

8. Top 30 Highest Average Rate and Total Votes per Episode Type and Saga. Sorted in Descending order based on Average Rate. 

```python
eight = naruto.groupby(by= ['Type','Saga']).agg({'Rate':'mean','Votes':sum}).sort_values(by= 'Rate', ascending = False).head(30)
eight
```

9. Top 20 Lowest Average Rate and Total Votes per Year and Saga. Sorted in Descending order based on Average Rate.

```python
nine = naruto.groupby(by=['Year_launch','Saga']).agg({'Rate':'mean','Votes':sum}).sort_values(by= 'Rate', ascending= False).tail(20)
nine
```

10. The Year and Episode Type that the Saga 1 Kazekage Rescue has the Highest Average Rates and Total number of Votes. Sorted in Descending Order.

```python
ten = naruto[naruto.Saga == 's1 Kazekage Rescue'].groupby(by=['Saga','Year_launch','Type']).agg({'Rate':'mean','Votes':sum}).sort_values(by= 'Rate', ascending =False)
ten
```
